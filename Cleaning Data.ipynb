```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("CAvideos.csv")
df
```

```python
df['trending_date'] = pd.to_datetime(df['trending_date'], format='%y.%d.%m')
df['publish_time'] = pd.to_datetime(df['publish_time'])
```

```python
df['publish_date'] = df['publish_time'].dt.date
df['publish_time_only'] = df['publish_time'].dt.time
```

```python
df['description'].fillna('', inplace=True)
```

```python
df['category_id'] = df['category_id'].astype(str)
```

```python
df.drop(columns=['publish_time'], inplace=True)
```

```python
columns_to_drop = ['video_id', 'thumbnail_link']
df_cleaned = df.drop(columns=columns_to_drop)
```

```python
df_cleaned = df_cleaned.drop_duplicates()
```

```python
df_cleaned['tags'] = df_cleaned['tags'].str.lower()
```

```python
numeric_columns = ['views', 'likes', 'dislikes', 'comment_count']
Q1 = df_cleaned[numeric_columns].quantile(0.25)
Q3 = df_cleaned[numeric_columns].quantile(0.75)
IQR = Q3 - Q1

outliers = (df_cleaned[numeric_columns] < (Q1 - 1.5 * IQR)) | (df_cleaned[numeric_columns] > (Q3 + 1.5 * IQR))
```

```python
df_cleaned = df_cleaned[~outliers.any(axis=1)]
```
